### Regressão Linear 

```python
#Importando as bibliotecas
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')

df_bef = pd.read_csv(r"c:\\Users\\gabya\Documents\ADA\\Curso - Data Science\\4. Estatística\\Projeto_EDA\\houses_to_rent_clean.csv")
print(df_bef.columns)
df = df_bef.copy()

## Pré-processamento

# OneHotEncoding em City, Animal, Furniture
col_one_hot_encoding = ['city', 'animal', 'furniture']

# Aplica one-hot encoding nas colunas selecionadas
df = pd.get_dummies(df, columns=col_one_hot_encoding)

# Removendo colunas desnecessárias
ind_col_drop = [1, 6, 8, 9, 10, 11, 12]
df.drop(df.columns[ind_col_drop], axis=1, inplace=True)

# Agora as colunas com os índices especificados foram removidas do DataFrame 'df'
# Visualização de Dados pós-processamento
print(df.info())

# Observando a correlação entre as variáveis
corr = df.corr()
print(corr)

## Regressão Linear
# Separando os dados
x = df.drop(['rent_amount'], axis = 1)
y = df.rent_amount

# Separação dos dados de testes e treino de forma aleatória
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

#Criando um Objeto de Regressão Linear
lr = LinearRegression()

#Treinando o Modelo
lr.fit(x_train,y_train)

#Verificando o resultado
r_sq = lr.score(x, y)
print('Coeficiente de Determinação (R²):', r_sq)

## Perfomance do Modelo
intercepto = lr.intercept_
print('Intercepto:', intercepto)

coeff_df = pd.DataFrame(lr.coef_,x.columns,columns=['Coefficient'])

## Erros
y_pred = lr.predict(x_test)
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# Gráfico
previsoes = lr.predict(x)
plt.figure(figsize=(10, 6))
plt.scatter(y, previsoes)
plt.xlabel("Valores Reais")
plt.ylabel("Previsões")
plt.title("Regressão Linear: Valores Reais vs Previsões")
plt.show()
```
